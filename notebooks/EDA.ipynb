```python
import pandas as pd
df = pd.read_csv("../data/raw/data.csv")
```

### Number of column,row and datatype

```python
row,col= df.shape
print('row:',row)
print('col:',col)


print('datatypes:')
df.dtypes
```

### central tendency, dispersion, and shape 

```python
import numpy as np

mean = np.mean(df['Amount'])
median = np.median(df['Amount'])
data_range = (np.max(df['Amount'])) - (np.min(df['Amount']))
std_dev = np.std(df['Amount'])
vari = np.var(df['Amount'])

print('mean:',mean)
print('median:',median)
print('range:',data_range)
print('variance:',vari)
print('standard deviation:',std_dev)


maximum = np.max(df["Amount"])
print('maximum:',maximum)
minimum = np.min(df["Amount"])
print('minimum:',minimum)
```

### Histogram for Amount,pricing and Fraud

```python
import matplotlib.pyplot as plt
import seaborn as sns


sns.histplot(df['Amount'],bins = 100,kde = True)
plt.title("Raw Amount Distribution")
plt.show()

sns.histplot(df['PricingStrategy'],bins = 100,kde = True)
plt.title("Raw PricingStrategy Distribution")
plt.show()

sns.histplot(df['FraudResult'],bins =2,kde = True)
plt.title("Raw FraudResult Distribution")
plt.show()

```

### Boxplot for  ProviderId,ProductId,ProductCategory,ChannelId

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='ProviderId', y='Amount', data=df)
plt.title("Transaction Amount by Provider")
plt.show()
```

```python
plt.figure(figsize=(14, 6))
sns.boxplot(x='ProductId', y='Amount', data=df)
plt.xticks(rotation=90)
plt.title("Transaction Amount by Product")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='ProductCategory', y='Amount', data=df)
plt.title("Transaction Amount by Product Category")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='ChannelId', y='Amount', data=df)
plt.title("Transaction Amount by Channel")
plt.show()
```

### correlation Analysis of numerical features

```python
numeric_col = ['Amount','PricingStrategy','FraudResult']



correlation = df[numeric_col].corr()

plt.figure(figsize=(8,6))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```
